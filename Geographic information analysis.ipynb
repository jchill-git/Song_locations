```python
def clean(string):
    clean_samp=string.replace('\'','')
    clean_samp=clean_samp.replace(', ',',')
    clean_samp=clean_samp.replace('\\','')
    return clean_samp
```

```python
def strToList(string):
    li = list(string.split(","))
    return li
```

```python
samp_loc_list=strToList(sample_locs)
samp_loc_list
```

```python
loc_list = []
```

```python
for index, song in geo_songs.iterrows():
    locs = geo_songs.at[index,'Locations'][1:-1]
    locs = clean(locs)
    song_locs = strToList(locs)
    loc_list = loc_list+song_locs
loc_list
```
