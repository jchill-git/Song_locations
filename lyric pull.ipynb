```python
top_songs = pd.read_csv('top_songs.csv', index_col=0)
```

```python
top_songs.head()
```

```python
artist = top_songs.at[0,'Artist']
title = top_songs.at[0,'Song']
song = LyricsGenius.search_song(artist, title)
```

```python
lyrics = song.lyrics
lyrics = lyrics.replace('\n','')
print(lyrics)
```

```python
top_songs.insert(4, 'Lyrics', 'words')
```

```python
top_songs.at[0,'Lyrics']=lyrics
top_songs.head()
```

```python
for index, row in top_songs.iterrows():
    artist = top_songs.at[index,'Artist']
    title = top_songs.at[index,'Song']
    song = LyricsGenius.search_song(title, artist)
    if song is None:
        print(title," has no lyrics")
    else:
        lyrics = song.lyrics
        lyrics = lyrics.replace('\n',' ')
        lyrics = lyrics.replace('Lyrics','')
        top_songs.at[index,'Lyrics']=lyrics
```
